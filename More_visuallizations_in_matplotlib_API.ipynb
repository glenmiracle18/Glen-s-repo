```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#There exist two methods of applyin visualizations to data sets
#We have the Global API and the OOP method which some times involves some looping
```

The Global API method

```python
x = np.arange(-10, 11)
```

```python
plt.figure(figsize=(12, 6))
plt.title('My Nice Plot')

plt.plot(x, x ** 2)
plt.plot(x, -1 * (x ** 2))
```

```python
plt.figure(figsize=(12, 6))
plt.title('My Nice Plot')

#first panel alocations
plt.subplot(1, 2, 1)  #rows, columns, panel selected
plt.plot(x, x ** 2)
plt.plot([0, 0, 0], [-10, 0, 100])
#this is for the above...
#cut the vertical line at point, zero, from zero to zero
plt.legend(['X^2', 'Vertical Line'])
plt.xlabel('X')
plt.ylabel('X Squared')

#second panel alocations
plt.subplot(1, 2, 2)  #rows, columns, 2nd panel
plt.plot(x, -1 * (x ** 2))
plt.plot([-10, 0, 10], [-50, -50, -50]) #Where the horizontal line should cut
#this is for the above... cut from -10, to +10 on the x axis and -50 on the y axis as well
plt.legend(['-X^2', 'Horizontal Line'])
plt.xlabel('X')
plt.ylabel('X squared')
```

Usint the OOP Interface method

```python
fig, axes = plt.subplots(figsize=(12, 6))
```

```python
axes.plot(
    x, (x ** 2), color='red', linewidth=3,
    marker='o', markersize=8, label='X^2')

axes.plot(
    x, -1 * (x ** 2), 'b--', label='-X^2')

axes.set_xlabel('X')
axes.set_ylabel('X Squared')

axes.set_title('My Nice Plot')
axes.legend()

fig
```

```python
fig, axes = plt.subplots(figsize=(12, 6))  #specifying the plotting frame

axes.plot(x, x + 0, linestyle='solid')
axes.plot(x, x + 1, linestyle='dashed')
axes.plot(x, x + 2, linestyle='dashdot')
axes.plot(x, x + 3, linestyle='dotted')

axes.set_title('My Nice Plot')
```

```python
fig, axes = plt.subplots(figsize=(12, 6))

#specifying the marker and the color respectively at the same time
axes.plot(x, x + 0, '-og', label='solid green')
axes.plot(x, x + 1, 'D', label='dashed cyan')
axes.plot(x, x + 2, '-.b', label='dashdot blue')
axes.plot(x, x + 3, ':r', label='dotted red')  #from here and below, I will practicing more or the markers asignment
axes.plot(x, x + 4, 's', label='dotted red')
axes.plot(x, x + 5, 'd', label='dotted red')
axes.plot(x, x + 6, '+', label='dotted red')
axes.plot(x, x + 7, '-->', label='dotted red')
axes.plot(x, x + 8, '<', label='dotted black')
#the '-og' is used to pre_announce that you will be using both marker and color at the same time

axes.set_title('My newest plot')
axes.legend() #used to creating a keybox for navigation around the plot
```

Learning how to work with Scatter plots

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (20 * np.random.rand(N))**2 #0 t0 15 point radii
```

```python
plt.figure(figsize=(14, 6))
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()
plt.show()
```

```python
fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(1,3,1)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel1')
plt.colorbar()

ax2 = fig.add_subplot(1,3,2)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel2')
plt.colorbar()

ax3 = fig.add_subplot(1,3,3)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel2')
plt.colorbar()

plt.show()
```

Working with Histograms

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
values = np.random.randn(500)

plt.subplots(figsize=(12, 6))

plt.hist(values, bins=30, alpha=0.8,
         histtype='bar', color='green',
         edgecolor='red')
#alpha parameter sets the transparency of the bars to 0.8 
#(i.e., slightly transparent).
plt.xlim(xmin=-5, xmax=5)
#plt.ylim(ymin=0, ymax=9)
plt.show()

#plt.xlim(xmin=-5, xmax=5) sets the limits of the x-axis to be from -5 to 5.
# This means that the histogram will only show the range of values from -5 to 5.
```

```python
fig.savefig('hist.png')
#how to save these snippets to your local machine
```

KDE (Kernel Density Estimation)

```python
from scipy import stats
density = stats.kde.gaussian_kde(values)
density
```

```python
plt.subplots(figsize=(12, 6))

values2 = np.linspace(min(values)-10, max(values)+10, 100)

plt.plot(values2, density(values2), color='#FF7F00')
plt.fill_between(values2, 0, density(values2), alpha=0.5, color='#FF7F00')
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

Combine plots, which one of the most important concepts to grasp.

```python
plt.subplots(figsize=(12, 6))

plt.hist(values, bins=50, alpha=0.8, density=1,
         histtype='bar', color='steelblue',
         edgecolor='green')

plt.plot(values2, density(values2), color='#FF7F00',
                          linewidth=3.0)
plt.xlim(xmin=-5, xmax=5)
plt.show()
```

Bar Plots

```python
Y = np.random.rand(1, 5)[0]
Y2 = np.random.rand(1, 5)[0]
```

```python
plt.figure(figsize=(12, 6))
barWidth = 0.5
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='#00b894')

plt.show()
```

```python
#Also can be stacked bars, and add a legend to the plot:

plt.figure(figsize=(12, 6))

barWidth = 0.5
plt.bar(np.arange(len(Y)), Y, width=barWidth, bottom=Y2, color='#00b894', label='Label Y')
plt.bar(np.arange(len(Y2)), Y2, width=barWidth, color='#e17055', label='Label y2')
#It is easier to spetify the position i.e 'bottom=y2', of the opposite bar plot
#on the line of the other bar plot. It just works better...

plt.legend()
plt.show()
```
